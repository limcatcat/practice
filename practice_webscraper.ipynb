```python
url = "https://coinmarketcap.com/"

result = requests.get(url)

doc3 = BeautifulSoup(result.text, "html.parser")

tbody = doc3.tbody
```

```python
print(tbody.prettify())
```

```python
trs = tbody.contents

print(trs[0].prettify())
```

```python
print(trs[0].parent.name)
```

```python
trs[0].find_all(string=re.compile("\$.*"))
```

```python
trs[0].find(class_="coin-item-name").string
```

```python
for tr in trs[0:10]:
    print(tr.find(class_="coin-item-name").string)
```

```python
prices = {}

for tr in trs[:10]:
    name, price = tr.contents[2:4]
    coin_name = name.find(class_="coin-item-name").string
    coin_price = price.string

    prices[coin_name] = coin_price
```

```python
print(prices)
```

```python
import pandas as pd


df_prices = pd.DataFrame(prices.items(), columns=['name', 'price'])


print(df_prices)
```

```python
df_prices['price'] = df_prices['price'].str.replace("$", "").str.replace(",", "")
```

```python
print(df_prices)
```

```python
df_prices['price'] = df_prices['price'].astype(float)
```

```python
df_prices_sorted = df_prices.sort_values(by="name", ascending=True)

print(df_prices_sorted)
```

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
url = "https://books.toscrape.com/catalogue/page-1.html"

result = requests.get(url)

doc = BeautifulSoup(result.text, "html.parser")
```

```python
print(doc.prettify())
```

```python
print(doc.title)
```

```python
products = doc.find_all("article", class_="product_pod")

len(products)
print(products)
```

```python
print(products[0].find_all("a")[1].attrs['title'])
```

```python
products[0]
```

```python
print(products[0].find(class_="product_price").p.string[2:])
```

```python
product_dict = {}

for product in products:
    title = product.find_all("a")[1].attrs['title']
    price = float(product.find(class_="product_price").p.string[2:])

    product_dict[title] = price


print(product_dict)
```
